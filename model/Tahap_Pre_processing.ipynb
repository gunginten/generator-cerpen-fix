```python
import pandas as pd

# Membaca dataset dari file (misalnya, CSV)
nama_file_input = '/content/drive/MyDrive/Deep Learning/Dataset/dataset_cerpen_genre.csv'
nama_file_output = '/content/drive/MyDrive/Deep Learning/Dataset/dataset_cerpen_genre_bersih.csv'

# Membaca dataset dari file ke DataFrame
df = pd.read_csv(nama_file_input)
df.info()

jumlah_kosong_total = df.isna().sum().sum()
print(jumlah_kosong_total)
```

# Menghandle Duplikasi Data

```python
duplikasi_dataframe = df.duplicated().sum()

print(duplikasi_dataframe)

#hasilnya bakal 0 karna kan id-nya beda-beda, caranya cek kolom
```

```python
duplikasi_kolom1 = df['title'].duplicated().sum()

print(duplikasi_kolom1)
```

```python
# Menghapus baris duplikat berdasarkan kolom 'title'
df_noduplicated = df.drop_duplicates(subset='title', keep='first')

df_noduplicated.count()
```

* Total Row pada Dataset adalah 21.789

# Memahami Dataset II
*

```python
df_noduplicated.head(100)
```

```python
df_noduplicated.tail(3)
```

```python
df_new = df_noduplicated.copy()
df_new['Teks_Cerpen'] = df_new['Teks_Cerpen'].str.split('Lolos moderasi pada:').str[1]

df_new.head(2)
```

```python
df_new['Teks_Cerpen'] = df_new['Teks_Cerpen'].apply(lambda x: ' '.join(x.split()[3:]))
df_new.head(2)
```

```python
df_new2=df_new.copy()
df_new2['Teks_Cerpen'] = df_new2['Teks_Cerpen'].str.replace('ADVERTISEMENT', '')
df_new2.tail(2)
```

```python
df_new2['Teks_Cerpen'] = df_new2['Teks_Cerpen'].str.split('Cerpen Karangan:').str[0]
df_new2.head(2)
```

```python
# Copy DataFrame
df_new3 = df_new2.copy()

df_new3['Teks_Cerpen'] = df_new3['Teks_Cerpen'].str.replace("at \d{1,2}:\d{2} [ap]m", " ")

#df_new3['Teks_Cerpen'] = df_new3['Teks_Cerpen'].str.strip()
df_new3.tail(2)
```

# Menghilangkan tanda \n dan \r

```python
# Menghilangkan \n dan \r dari seluruh dataset
df_after = df_new2.applymap(lambda x: x.replace('\n', '').replace('\r', '') if isinstance(x, str) else x)

print("\nDataFrame setelah menghilangkan \\n dan \\r:")
df_after.tail()
```

# Menghapus Kolom yang Tidak Diperlukan

```python
column_drop=['Unnamed: 0', 'url']
df_after=df_after.drop(column_drop,axis=1)
df_after.head()
```

# Menyimpan dalam CSV baru

```python
df_after.to_csv(nama_file_output)
```
