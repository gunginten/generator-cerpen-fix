```python
import pandas as pd
import glob
import os

file_path = "/content/drive/MyDrive/Deep Learning/DatasetGenre/DatasetGenre"
os.chdir(file_path)  # Change the current working directory to the specified path
file_list = glob.glob('*.{}'.format('csv'))
file_list
```

```python
dataset_cerpen =pd.DataFrame()

for dataset in file_list:
  df_temp =pd.read_csv(dataset)
  dataset_cerpen=dataset_cerpen.append(df_temp, ignore_index=True)
dataset_cerpen
```

```python
dataset_cerpen.to_csv('/content/drive/MyDrive/Deep Learning/Dataset/dataset_cerpen_genre.csv')
```
